# Probability for limit-order execution

```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)


def f_Q(Q,alpha):
    """
    Power law density for market order sizes
    """
    return Q**(-1-alpha)

def cumul_log_exec_prob(delta,alpha,K):
    """
    Cumulative probability for limit order execution at depth level
    delta assuming logarithmic market impact.
    """
    return np.exp(-alpha*K*delta)/alpha


```

```python
# limit order placement distances from current mid-point
deltas = np.arange(0.01,2,0.01)

# limit order book parameters
alpha_1 = 1.00
alpha_2 = 1.25
alpha_3 = 1.50
K       = 1.0

f = plt.figure(figsize=(10,10))
plt.plot(deltas,cumul_log_exec_prob(deltas,alpha_1,K),color='blue',ls=':',lw=2)
plt.plot(deltas,cumul_log_exec_prob(deltas,alpha_2,K),color='blue',ls=':',lw=2)
plt.plot(deltas,cumul_log_exec_prob(deltas,alpha_3,K),color='blue',ls=':',lw=2)

plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
#plt.title('Limit-order execution probability',fontsize=20)
plt.ylabel('$P(\Delta S \geq \delta)$',fontsize=24)
plt.xlabel('$ \delta $',fontsize=24)
plt.show()
```

```python
import seaborn as sns
import matplotlib.ticker as ticker

alphas = np.linspace(1.0,2.5,150)
P_d    = np.zeros((len(alphas),len(deltas)))
P_c    = np.zeros((len(alphas),len(deltas)))

# Conditional execution densities
for i,alpha in enumerate(alphas):
    P_d[i,:] = cumul_log_exec_prob(deltas,alpha,K)

# Conditional execution cumulative probabilities
for i,alpha in enumerate(alphas):
    P_c[i,:] = cumul_log_exec_prob(deltas,alpha,K)

ts      = 30
fs      = 20
fs_cbar = 15
x_label = '$\delta $'
y_label = r'$\alpha$'

a, d    = np.meshgrid(deltas, alphas)
fig, ax = plt.subplots(figsize=(12,10))
im      = ax.pcolormesh(a, d, P_c, cmap='jet')
cbar    = fig.colorbar(im)
cbar.ax.tick_params(labelsize=fs_cbar)

plt.xlabel(x_label,fontsize=ts)
plt.ylabel(y_label,fontsize=ts)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs)
ax.axis('tight')
plt.show()
```
