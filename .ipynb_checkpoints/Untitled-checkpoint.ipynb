# Probability for limit-order execution

```python
import numpy as np
import matplotlib.pyplot as plt

def f_Q(Q,alpha):
    """
    Power law density for market order sizes
    """
    return Q**(-1-alpha)

def cumul_log_exec_prob(delta,alpha,K):
    """
    Cumulative probability for limit order execution at depth level
    delta assuming logarithmic market impact.
    """
    return np.exp(-alpha*K*delta)/alpha


```

```python
# limit order placement distances from current mid-point
deltas = np.arange(0.01,2,0.01)

# limit order book parameters
alpha_1 = 1.00
alpha_2 = 1.25
alpha_3 = 1.50
K       = 1.0

f = plt.figure(figsize=(10,10))
plt.plot(deltas,cumul_log_exec_prob(deltas,alpha_1,K),color='blue',ls=':',lw=2)
plt.plot(deltas,cumul_log_exec_prob(deltas,alpha_2,K),color='blue',ls=':',lw=2)
plt.plot(deltas,cumul_log_exec_prob(deltas,alpha_3,K),color='blue',ls=':',lw=2)

plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.title('Limit-order execution probability',fontsize=20)
plt.ylabel('$P(\Delta S \geq \delta)$',fontsize=24)
plt.show()
```

```python
import seaborn as sns

alphas = np.linspace(1.0,2.5,10)

P      = np.zeros((len(alphas),len(deltas)))

print(P)

for i,alpha in enumerate(alphas):
    
    P[i,:] = cumul_log_exec_prob(deltas,alpha,K)

f = plt.figure(figsize=(12,10))
ax = sns.heatmap(P, vmin=0, vmax=1,cmap="jet",yticklabels=np.round(alphas,0))
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.ylabel(r'$\alpha$',fontsize=24)
plt.show()
```
